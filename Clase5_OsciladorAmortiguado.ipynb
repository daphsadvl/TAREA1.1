```python

import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint 
%matplotlib inline
```

```python
k= 3.0 # constante muelle
m= 1.0 # masa
B= .5 # constante de amortiguacion

def armonico (variables, t):
    x, y= variables
    return [y, -k * x / m-B / m * y]
inicial= [ 0.6,0.4]# vector de posicio inicial  y velocidad inical 
# condiciones inicales  x(t=0)=0.6 [m]  y (t=0)=0.4 [m/s]

#tiempo =np.linspace(0,15) # dominio temporal de 0 a 15
tiempo = np.arange(0, 20, .01)

resultado=odeint(armonico, inicial, tiempo)
# el sistema se resulve con 
#odeint(sistema, condiciones iniciales , rango donde graficaremos)
```

```python
xx, yy = resultado.T #extraer posicion y velocidad 
```

```python
import matplotlib as mpl
label_size = 14
mpl.rcParams['xtick.labelsize'] = label_size 
mpl.rcParams['ytick.labelsize'] = label_size
```

```python
plt.plot(tiempo, xx, c = 'r', label="Posicion") 
plt.plot(tiempo, yy, c = 'k', label="Velocidad")
plt.legend(loc = 'best', prop={'size': 14})    
plt.xlabel('tiempo', fontsize = 14) 
plt.show()
```

```python

omega0 = k/m
Gamma = B/(2*m)
```

```python
omega0**2, Gamma**2
```

```python
omega0**2 > Gamma**2
```

```python
plt.figure(figsize = (6,6))
plt.scatter(xx, yy/omega0, lw = 0, s = 3, cmap = 'viridis', c = xx)
plt.show()
```
