```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.arange(0,5,.1)
y=x**2
plt.plot(x,y)
```

```python
import matplotlib as  mlp
label_size=14
mpl.rcParams["xtick.labelsiza"]= label_size
mpl.rcParams["ytick.labelsiza"]= label_size

```

```python
A=2
W=0.05
B=0.1
t= np.linspace(0,500,1000)
plt.figure(figsize=(7,4))

plt.plot(t, W *np.cos(W *t)+ .1 * np.sin(W*t), "rv", lw=1, ms =4,
         label = "$x(t)$")
plt.plot(t, B*W*np.cos(W*t)-A*W * np.sin(W*t), "bo-", lw=1, ms=7,
         label=r"$\dot{x(t)$")

plt.xlabel("$t$", fontsize = 20)
plt.show()
    
```

```python
from ipywidgets import *
```

```python

def masa_resorte(t = 0):
    fig = plt.figure()
    ax = fig.add_subplot(1, 1, 1)
    ax.plot(.5 * np.cos(.5 * t) + .1 * np.sin(.5 * t), [0],  'ko', ms = 10)
    ax.set_xlim(xmin = -0.6, xmax = .6)
    ax.axvline(x=0, color = 'r')
    ax.axhline(y=0, color = 'grey', lw = 1)
    fig.canvas.draw()

interact(masa_resorte, t = (0, 50,.01));
```
