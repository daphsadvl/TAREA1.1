```python
file_amazon="precios/AMZN.csv"
file_microsoft="precios/MSFT.csv"
file_nvidia="precios/NVDA.csv"

df_amazon=pd.read_csv (file_amazon, index_col="Date", usecols=["Date", "Adj Close"])

df_microsoft=pd.read_csv (file_microsoft, index_col="Date", usecols=["Date", "Adj Close"])

df_nvidia=pd.read_csv (file_nvidia, index_col="Date", usecols=["Date", "Adj Close"])

```

```python
closes = pd.DataFrame(index=df_amazon.index, columns=["APPL", "AMZN", "MSFT", "NVDA"])
closes.index.name="Date"
closes["APPL"] = df_apple
closes["AMZN"] = df_amazon
closes["MSFT"] = df_microsoft
closes["NVDA"] = df_nvidia
closes
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
closes.plot(figsize=(8,6));
```
