```python
%matplotlib inline 

import numpy as np
import matplotlib.pyplot as plt
import random 
```

```python
%%timeit
def walker():
    global x,y
    eleccion=random.uniform(0,1)
    if eleccion < 0.40:
        x+=1
    elif eleccion < 0.60:
        x-=1
    elif eleccion < 0.8 :
        y+=1
    else:
        y-=1
    return [x,y]
x=0 #iniciar el caminador
y=0

Pos= np.zeros((2,1000))

for i in np.arange(0, len(Pos [0]),1):
    Pos[:,i]= walker()
    
plt.plot(Pos[0,:],Pos[1,:], "bo-")
plt.plot(Pos[0,-1],Pos[1,-1], "ro-")
```

```python
%%timeit
inicio= np.array([[0],[0]])

elements = np.array([[1,1,0,0],[0,0,1,-1]])

probabilities= [0.4,0.2,0.2,0.2]

pasos = elements[:, np.random.choice(elements.shape[1], 1000, p= probabilities)]
pasos= np.concatenate((inicio, pasos), axis=1 )

Pos= np.cumsum(pasos, axis=1)
plt.plot(Pos[0,:], Pos[1,:], "bo-")
```
